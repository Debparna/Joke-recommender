```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%pylab inline

df = pd.read_csv("data/UserRatings1.csv")
print(df.shape)
df.head()
```

```python
df = df.melt(id_vars="JokeId",var_name = "User", value_name="rating")
df.head()
```

```python
df["mean_joke_rating"] = df.groupby("JokeId")["rating"].transform("mean")
df["mean_user_rating"] = df.groupby("User")["rating"].transform("mean")
df["user-count"] = df.groupby("User")["rating"].transform("count")
df["joke-count"] = df.groupby("JokeId")["rating"].transform("count")
df.describe()
```

```python
df.head()
```

```python
df = df.merge(pd.read_csv("data/JokeText.csv"), on="JokeId")
df.tail()
```

```python
df.rating.mean()
```

```python
df.rating.median()
```

```python
df3 = df.drop_duplicates(subset=["JokeId"]).loc[:, ["JokeId", "mean_joke_rating", "joke-count", "JokeText"]].sort_values("mean_joke_rating", ascending= False)
df3.head()
```

```python
# The funniest
for i in list(df3.head().JokeText):
    print(i)
    print("-------------------------------------")
```
